```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yeses=['yes','y','YES','Yes']
nos=['no','n','NO','No']

s = input("Do you want grid lines (yes/no) ")
t = input("Do you want tick labels (yes/no) ")
m = "gist_earth"

image = plt.imread('stinkbug.png')

fig, ax = plt.subplots()

ax.imshow(image, cmap=m)

if t in nos:
    ax.set_xticklabels([])  # supply an empty list
    ax.set_yticklabels([])
    ax.tick_params(axis='both', length=0.0)

if s in yeses:
    ax.grid(True)
else:
    ax.grid(False)


plt.show()
```
